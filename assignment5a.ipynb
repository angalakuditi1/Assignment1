Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class is a base class for exceptions raised when an operation or function attempts to access an element in a collection like a list, dictionary, or tuple, but the value cannot be found. It is a subclass of the built-in Exception class.

KeyError is something happens when we try to access a dictionary with key that doesn't exist
Where as IndexError happens when try to access a index which is out of range

```python
Hero = {'name': 'mahesh', 'age': 49}

try:
    print(Hero['gender']) 
except KeyError as e:
    print(f"KeyError: {e}")
```

```python
my_list = [1, 2, 3, 4 ]

try:
    print(my_list[7])  
except IndexError as e:
    print(f"IndexError: {e}")
```
