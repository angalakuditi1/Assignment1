Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence. Provide a sample problem and solution.

```python
#A companys email filter flags emails as spam if they contain the word "GIFT". Given:

#As,per bayees formula 

PS = 0.30  # Probability of spam
PW_S = 0.80  # Probability of "GOOD" appearing in spam email
PH = 1 - PS  # Probability of non-spam email
PW_H = 0.10  # Probability of "GOOD" appearing in a non-spam email

# Compute total probability of "GOOD" appearing 
PW = (PW_S * PS) + (PW_H * PH)

# Apply Bayes' Theorem
PS_W = (PW_S * PS) / PW

# Print the final result
print(f"If an email contains 'lottery', the probability that it is spam is: {PS_W:.2%}")
```

Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

```python
import scipy.stats as stat
import numpy as np
```

```python
import scipy.stats as stats
import numpy as np

sample_data = np.random.normal(loc=6, scale=2.5, size=50) 
hypothesized_mean = 0

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(sample_data, hypothesized_mean, alternative='greater')

# Set significance level
alpha = 0.05

# Output results
print(f"T-Statistic: {t_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")
if p_value < alpha:
    print("Reject the null hypothesis: The drug is significantly effective.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that the drug is effective.")
```

Q12. A researcher is testing the effectiveness of two different teaching methods on student performance.
Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82
with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a
significant difference in student performance using a t-test with a significance level of 0.01.

Null hypothesis - HO - Reject the null hypothesis: There is a significant difference in student performance.
Alternative hypothesis - H1 - Failed to reject the null hypothesis: There is significant difference in student

```python
import numpy as np
import scipy.stats as stats

import scipy.stats as stats

mean_A = 85
std_A = 6
n_A = 30  # Assume sample size

mean_B = 82
std_B = 5
n_B = 30  # Assume sample size


t_stat, p_value = stats.ttest_ind_from_stats(mean1=mean_A, std1=std_A, nobs1=n_A,
                                             mean2=mean_B, std2=std_B, nobs2=n_B,
                                             equal_var=True)
alpha = 0.01

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in student performance.")
else:
    print("Fail to reject the null hypothesis: No significant difference in student performance.")
```

Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average
reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to
determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.

```python
import scipy.stats as stats
import numpy as np

caffeine_group = np.random.normal(0.25, 0.05, 30)  
control_group = np.random.normal(0.30, 0.05, 30)  

t_statistic, p_value = stats.ttest_ind(caffeine_group, control_group, equal_var=True)

print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")

alpha = 0.10
if p_value < alpha:
    print("Reject the null hypothesis: Caffeine has a significant effect on reaction time.")
else:
    print("Fail to reject the null hypothesis: No significant effect of caffeine on reaction time.")
```
