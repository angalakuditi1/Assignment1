Q2. Give advantages and disadvantages of using API.

pros
APIs enable automation of repetitive tasks, reducing manual effort and errors.
Developers can use predefined methods instead of building features from scratch.
Many APIs follow standard protocols  that promote consistency and ease of use.

cons
APIs are vulnerable to security threats like unauthorized access, data breaches, and injection attacks if not properly secured.
Many APIs come with usage limits and costs for higher tiers, which can be expensive for large-scale use.

Q4. Explain REST and SOAP Architecture. Mention shortcomings of SOAP.

SOAP  and REST  are two architectures commonly used in web services to enable communication between client and server,SOAP is a protocol that defines a strict set of rules for message structure and communication.

REST is an architectural style based on standard HTTP methods, making it simpler and more flexible than SOAP.

SOAP is more complex due to its strict standards, requiring XML-based messages and detailed configurations,SOAP has a steeper learning curve and less compatibility with web-based technologies like browsers.

Q5. Differentiate between REST and SOAP.

SOAP is a protocol with strict standards for message formats, security, and communication,it uses XML for message formatting, which can be more complex and verbose,relies on request-response messages using specific operations (like get, post, etc.), typically over HTTP or other protocols,provides more built-in security features (e.g., WS-Security), making it suitable for enterprise-level applications.


REST is an architectural style based on HTTP, focusing on simplicity and stateless operations, relies on underlying protocols like HTTPS for security, with fewer built-in security features,uses standard HTTP methods (GET, POST, PUT, DELETE) and is more flexible and scalable,it primarily uses JSON, though it can support XML, plain text, or other formats, and is generally simpler and lighter.
