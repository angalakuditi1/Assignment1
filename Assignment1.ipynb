#Q7In programming, mutable and immutable refer to whether or not a data type can be modified after it is created.

Mutable objects can be changed after their creation,
meaning their contents can be altered without changing their identity
Examples in Python include lists, dictionaries, and sets.

Immutable objects cannot be changed after they are created. 
Any operation that tries to modify the object will result in the creation of a new object.
Examples include strings, tuples, and numbers (integers, floats).
