```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
ne = load_iris()
```

```python
x = ne.data[:,:2]
y = ne.target
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.30)
```

```python
x_test.shape
```

```python
svm = SVC(kernel='poly')
```

```python
svm.fit(x_train,y_train)
```

```python
y_pred = svm.predict(x_test)
```

```python
accuracy = accuracy_score(y_pred,y_test)
print(f'the accuracy is {accuracy}')
```

```python
#plotting the decision trees
x_min,x_max= x[:,0].min()-1,x[:,0].min()+1
y_min,y_max= x[:, 1].min() - 1, x[:, 1].max() + 1
```

```python
x1,y1 = np.meshgrid(np.arange(x_min,x_max,0.01),np.arange(y_min,y_max,0.01))
```

```python
z = svm.predict(np.c_[x1.ravel(),y1.ravel()])
```

```python
z = z.reshape(x1.shape)
```

```python
plt.contourf(x1, y1, z, cmap=plt.cm.coolwarm, alpha=0.8)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.coolwarm, marker='o', edgecolors='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM with Polynomial Kernel (Degree 3)')
plt.show()
```
