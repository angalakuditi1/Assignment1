Q1.What is multiprocessing in python? Why is it useful?

Multiprocessing in Python is a module that allows you to create and manage multiple processes in a program. Processes are independent units of execution that run in separate memory spaces, unlike threads that share the same memory space.

Each process has its own memory space, which reduces the chance of conflicts,Tasks that can run independently can be distributed across processes.

Q3. Write a python code to create a process using the multiprocessing module.

```python
import multiprocessing
import time
import sys

def cooking():
    print("I am cooking")
    time.sleep(1)
    print("Cooking done!")

def chopping():
    print("I am chopping")
    time.sleep(2)
    print("Chopping done!")

def cleaning():
    print("I am cleaning")
    time.sleep(3)
    print("Cleaning done!")

if __name__ == "__main__":
    work1 = multiprocessing.Process(target=cooking)
    work2 = multiprocessing.Process(target=chopping)
    work3 = multiprocessing.Process(target=cleaning)

    work1.start()
    work2.start()
    work3.start()

    work1.join()
    work2.join()
    work3.join()

    print("All tasks completed!")
```

Q4. What is a multiprocessing pool in python? Why is it used?

A multiprocessing pool in Python is a class provided by the multiprocessing module that allows you to manage a pool of worker processes. These workers can execute tasks in parallel, making it easy to distribute workload across multiple processes.

Q5. How can we create a pool of worker processes in python using the multiprocessing module?

In Python, you can create a pool of worker processes using the multiprocessing.Pool class. A pool allows you to manage a fixed number of worker processes to which tasks can be assigned. The Pool automatically distributes these tasks among the workers and collects the results.

```python
from multiprocessing import Pool

def square_number(n):
    return n * n

if __name__ == "__main__":
    numbers = [1, 2, 3, 4, 5] 
    with Pool(processes=4) as pool:
         results = pool.map(square_number, numbers)
    print(f"Squared results: {results}")
```
