Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

there are many plots that can be plot using seaborn library one of the five are
scatterplot
Use: Visualizes the relationship between two continuous variables.

Histogram
Use: Shows the distribution of a single variable.

Box Plot 
Use: Summarizes data using the median, quartiles, and potential outliers.

Heatmap
Use: Displays data in a tabular format with colors representing values.

Bar Plot 
Use: Displays the mean of a categorical variable with error bars.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import numpy as np
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(x ='timepoint',y = 'signal',data = fmri ,style='region',hue= 'event')
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
dia = sns.load_dataset('diamonds')
```

```python
dia
```

```python
sns.histplot(x = 'price',data= dia,hue = 'cut')
```
