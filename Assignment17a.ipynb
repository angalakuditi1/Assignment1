```python
def area(x,std,mean):
     coefficient = 1 / (std * np.sqrt(2 * np.pi))
     exponent = np.exp(-0.5 * ((x - mean) / std) ** 2)
     return coefficient * exponent

x = 4
std = 1
mean = 0 

pdf_values = area(x,std,mean)
print(f'the probability density function of a normal distribution is {pdf_values}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

nw =  np.random.binomial(1000,0.4, 1000)

plt.hist(nw,bins = 60,color = 'yellow',edgecolor = "blue")
plt.title('Histogram of 1000 Samples from Binomial Distribution (n=1000, p=0.4)')
plt.xlabel('number of sucesss')
plt.ylabel('frequency')
```

```python
import math

def cumulative(k,mean):
     cdf = 0 
     for x in range(k + 1):
        poisson_prob = (mean**x * math.exp(-mean)) / math.factorial(x)
        cdf += poisson_prob
     return cdf

cumulative(12,18)
```

```python
lam = 5
nw = np.random.poisson(lam = lam, size = 1000)

print('the mean of the sample is:',np.mean(nw))
print('the variance of the sample is :',np.var(nw))
```
