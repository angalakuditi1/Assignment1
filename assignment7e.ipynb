Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are like blank forms for creating web pages. They are mostly written in HTML but have some special placeholders (like empty boxes) where we can insert data.

For example, if we  want to create a webpage that says "Hello, [name]!" but the name should change depending on who visits the page. Instead of writing different HTML for each name, you can use a template where [name] is a placeholder.

The render_template() function is what fills these placeholders with real data and sends the final webpage to the user's browser.

```python
from flask import Flask, jsonify, request

app = Flask(__name__)

@app.route('/api/greet', methods=['GET'])
def greet():
    name = request.args.get('name', 'Guest')
    return jsonify(message=f"Hello, {name}!")

if __name__ == '__main__':
    app.run(debug=True)
```
