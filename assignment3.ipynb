Q5. Create a generator function for prime numbers less than 1000. Use the next() method to print the
first 20 prime numbers.

```python
def prime_generator():
    for n in range(2, 1000):
        is_prime = True
        for i in range(2, n):
            if n % i == 0:
                is_prime = False
                break
        if is_prime:
            yield n

prime_gen = prime_generator()

for _ in range(20):
    print(next(prime_gen))
```
