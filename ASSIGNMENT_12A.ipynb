Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.graph_objects as go
```

```python
tit = sns.load_dataset('titanic')
```

```python
fig = go.Figure()
```

```python
tit
```

```python
fig.add_trace(go.Scatter(x=tit.age,y= tit.fare,mode = 'markers'))
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

displot is a function from the Seaborn library in Python, used for creating distribution plots. It is a versatile function that can display various types of distribution plots, such as histograms, kernel density estimates, and ECDFs.

```python
import plotly.express as px

import numpy as np
data = np.random.normal(0, 1, 1000)

fig = px.histogram(data, nbins=30, title="Distribution Plot")
fig.show()
```
