Q2. What is the objective function of a linear SVM?

The objective function of a linear SVM (Support Vector Machine) aims to find the best straight line (or hyperplane) that separates two classes of data with the maximum margin—meaning, it wants to keep the dividing line as far away as possible from both classes, so it’s more confident about its decisions.

This function has two goals:

Maximize the margin between the two classes (i.e., make the line clear and confident).

Minimize classification errors using a penalty if some points are on the wrong side or too close to the line (this is where a regularization term helps).

In math terms, it tries to minimize:
½ ||w||² + C × (sum of errors)
Here:

||w||² keeps the margin wide.

C controls how much penalty we give for mistakes.

In short, it balances accuracy and confidence.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df['target'] = dataset.target
```

```python
#Dependent and Independent Features
X = df.iloc[:,:-1]
y = df.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
svm = SVC(kernel = 'linear')
```

```python
svm.fit(X_train,y_train)
```

```python
y_pred = svm.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
cf = confusion_matrix(y_test,y_pred)
sns.heatmap(cf,annot=True)
```

```python
# Plot the decision boundaries of the trained model using two of the features
df_train = pd.concat([X_train,y_train],axis=1)
sns.scatterplot(data = df_train, x = 'sepal length (cm)',y='sepal width (cm)',hue = y_train)
plt.title('Sepal Length vs Sepal width')
plt.show()
```
