Q2. What is the V-measure in clustering evaluation? How is it related to homogeneity and completeness?

V-measure is a way to evaluate how good a clustering algorithm is, by comparing its results to the actual (true) groupings of data.

In simple terms, it checks two things:

Homogeneity – Are all the items in each predicted cluster mostly from the same actual category? (Good clustering keeps similar things together.)

Completeness – Are all items of a true category assigned to the same cluster? (No splitting up of a category across many clusters.)

V-measure combines both of these into a single score using their harmonic mean. This means the score is high only if both homogeneity and completeness are high. It ranges from 0 to 1, where 1 means perfect clustering.
