Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results.

```python
import numpy as np
import scipy.stats as stats

#we have the data

sample_mean = 50
sample_std = 5
confidence = 0.95
sample_size = 30 # taking an assumption as we dont have sample

df = sample_size-1

t_value = stats.t.ppf((confidence+1)/2,df)

margin_of_error = t_value * (sample_std / np.sqrt(sample_size))

upper_bound = sample_mean + margin_of_error
lower_bound = sample_mean - margin_of_error

print(f"the confidence interval for the given data is {upper_bound:.2f},{lower_bound:.2f}")
```

```python
import scipy.stats as stats

# Step 1: Observed data (assumed counts in a bag)
observed = [25, 30, 15, 10, 8, 12]  # Blue, Orange, Green, Yellow, Red, Brown

# Step 2: Expected proportions (given in the question)
expected_proportions = [0.2, 0.2, 0.2, 0.1, 0.1, 0.2]  # Based on percentage

# Step 3: Calculate expected counts based on total observed data
total_count = sum(observed)  # Total M&Ms counted
expected = [p * total_count for p in expected_proportions]

# Step 4: Perform Chi-Square Goodness of Fit Test
chi2_stat, p_value = stats.chisquare(observed, expected)

# Step 5: Print results
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 6: Decision making at α = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H₀: The observed color distribution is significantly different from expected.")
else:
    print("Fail to reject H₀: The observed color distribution matches the expected distribution.")
```

```python
import scipy.stats as stats
import numpy as np

# Given values
mean = 75
std_dev = 12
sample_size = 30  # Assuming a sample size of 30
confidence_level = 0.90

# Z-score for 90% confidence interval
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Margin of error
margin_of_error = z_score * (std_dev / np.sqrt(sample_size))

# Confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"90% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# Define the contingency table as a 2x2 numpy array
observed = np.array([[60,140],  # smoker
                     [30,170]]) # no smoker
print(observed)

# Perform the chi-square test
chi2, p_value, dof, expected = chi2_contingency(observed)

# Print the results
print(f'p_value is {p_value}')
print(f'chi2 is {chi2}')
print(f'expected is {expected}')

# Decision rule
if p_value < alpha:
    print("Reject the null hypothesis at 0.05. There is a significant association between smoking status and lung cancer diagnosis")
else:
    print("Fail to reject the null hypothesis at 0.05. There is no significant association between smoking status and lung cancer diagnosis.")
```
